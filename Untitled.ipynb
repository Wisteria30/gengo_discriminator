```python
# coding: utf-8

#回帰分析のプログラム
#作成日：2019年1月15日

import numpy as np
from matplotlib import pylab
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline


if __name__ == "__main__":
    # データのファイル入力
    #inputData = np.array([ [3.0, 10.0], [8.0, 5.0], [6.0, 6.0], [8.0, 7.0], [15.0, 2.0] ])  #説明変数が1つの場合
    inputData = np.array([ [0.0, 0.0, 2.0], [2.0, 3.0, 5.0], [6.0, 5.0, 6.0], [-2.0, -1.0, 1.0], [5.0, 4.0, -1.0] ])  #説明変数が1つの場合
    print("inputData = \n", inputData)    # 入力データの確認


    # データ数、次元数の取得
    n = len(inputData)    # データ数
    p = inputData.shape[1] - 1    # 次元数
    print("データ数は %d, 次元数は %d" % (n, p) )    #データ数と次元数の確認

    # データの整形(Start) -----------------
    X1 = np.ones( (n, 1) )    # n行1列のベクトルを生成（成分はすべて1）
    Xdata = inputData[:, 0:p]    # 入力データの説明変数のみを取り出し
    X = np.c_[X1, Xdata]     # 配列の結合
    Y = inputData[:, -1].reshape( (n, 1) )     # 入力データの目的変数のみを取り出し
    #print("X1 = ", X1)
    #print("Xdata = ", Xdata)
    #print("X = ", X)
    #print("Y = ", Y)
    # データの整形(End) -----------------


    # 重みベクトルの更新(Start) ---------
    W = np.dot(np.dot( np.linalg.inv(np.dot(X.T, X)), X.T), Y)
    print("重みベクトルWの更新結果：-----------")
    print("W = \n", W)
    # 重みベクトルの更新(End) ---------


    #----------------------------------------------
    #これ以降は表示関係
    #----------------------------------------------

    #日本語テキスト描画の準備
    #fp = FontProperties(fname=r'/usr/share/fonts/opentype/ipafont-gothic/ipag.ttf') # IPAフォント


    if p == 1:
      # 回帰式を描画するために最大値・最小値を取得
      minX = np.min(X[:,1])
      maxX = np.max(X[:,1])
      minY = np.min(Y)
      maxY = np.max(Y)
      print("X の最小値は ", minX)
      print("X の最大値は ", maxX)
      print("Y の最小値は ", minY)
      print("Y の最大値は ", maxY)


      # 回帰式を描画する準備(Start) ---------
      x = range( int(np.floor(minX)) - 1, int(np.ceil(maxX)) + 2)
      y = [W[0] + W[1] * k for k in x ]
      # 回帰式を描画する準備(End) ---------

      #図のオプションなど
      pylab.grid(True)   #gridの表示
      pylab.axis([int(np.floor(minX)) - 1, int(np.ceil(maxX)) + 1, int(np.floor(minY)) - 1, int(np.ceil(maxY)) + 1])   #描画範囲の指定

      #結果の表示
      pylab.scatter(Xdata, Y, color='k', marker='x', label='data point', s=100)
      pylab.plot(x, y, 'b', label='regression model', linewidth=1.5)

      #pylab.legend(loc='upper right')   #凡例表示

    if p == 2:
      # 回帰式を描画するために最大値・最小値を取得
      minX1 = np.min(X[:,1])
      maxX1 = np.max(X[:,1])
      minX2 = np.min(X[:,2])
      maxX2 = np.max(X[:,2])
      minY = np.min(Y)
      maxY = np.max(Y)
      print("X1 の最小値は ", minX1)
      print("X1 の最大値は ", maxX1)
      print("X2 の最小値は ", minX2)
      print("X2 の最大値は ", maxX2)
      print("Y の最小値は ", minY)
      print("Y の最大値は ", maxY)

      xmesh, ymesh = np.meshgrid(np.linspace(minX1, maxX1, 20),
                                                          np.linspace(minX2, maxX2, 20))
      zmesh = (W[0] + W[1] * xmesh.ravel() + W[2] * ymesh.ravel()).reshape(xmesh.shape)
      fig = pylab.figure()
      ax = fig.add_subplot(111, projection = '3d')
      ax.scatter(X[ :, 1], X[ :, 2], Y, color='k')
      ax.plot_wireframe(xmesh, ymesh, zmesh, color='b')


    pylab.show()   #結果の表示
```
